```python
import baltic as bt
import pandas as pd
import numpy as np
import json
import os
import matplotlib.pyplot as plt
import itertools
```

```python
def load_tree(gene):
    path = f'./auspice/concat.json'
    mytree, mymeta = bt.loadJSON(path)
    return(mytree)
```

```python
all_genes_data = {}

list_of_genes = ["HA", "PB1", "PB2","PA","NP","NA","NS"]

path = f'./auspice/concat.json'
mytree, mymeta = bt.loadJSON(path)

for k in mytree.Objects:
    
    if k != mytree.root and (k.branchType == "node" or k.branchType == "leaf"):
        
        name = k.traits["name"]
        length = k.traits["num_date"] - k.parent.traits["num_date"]
        
        nt_mutations = [mut for mut in k.traits["branch_attrs"]["mutations"].get("nuc", []) if "-" not in mut and "N" not in mut]
        
        aa_mutations = []
        
        for gene in list_of_genes:
            aa_mutations.extend([mut for mut in k.traits["branch_attrs"]["mutations"].get(gene, []) 
                if "-" not in mut and "X" not in mut
                ])
          
        # print(name, aa_mutations)
        all_nt = len(nt_mutations)

        nonsyn = len(aa_mutations)
        syn = all_nt - nonsyn

        all_genes_data[name] = (length, all_nt, nonsyn, syn)
```

```python
x_vals = []
y_vals = []  
names = []

for key, (length, all_nt, nonsyn, syn) in all_genes_data.items():
    if length > 1:
        x = syn / length
        y = nonsyn / length
        x_vals.append(x)
        y_vals.append(y)

# Convert to numpy arrays
x_vals = np.array(x_vals)
y_vals = np.array(y_vals)

# Compute percentiles
x_95 = np.percentile(x_vals, 95)
y_95 = np.percentile(y_vals, 95)

# Assign colors based on conditions
for key, (length, all_nt, nonsyn, syn) in all_genes_data.items():
    if length > 1:
        x = syn / length
        y = nonsyn / length
        
        if y > y_95 and x < x_95:
            names.append(key)


# Plot
plt.figure(figsize=(8, 6))
plt.scatter(x_vals, y_vals, c="black", alpha=0.7, edgecolors='k')
plt.axvline(x_95, color='blue', linestyle='dashed', linewidth=1)
plt.axhline(y_95, color='red', linestyle='dashed', linewidth=1)
plt.xlabel("syn subs / branch length")
plt.ylabel("nonsyn subs / branch length")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
