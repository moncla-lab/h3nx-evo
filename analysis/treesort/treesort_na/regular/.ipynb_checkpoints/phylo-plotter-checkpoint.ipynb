```python
from Bio import Phylo
import matplotlib
import matplotlib.pyplot as plt

tree = Phylo.read("trees/modified_no-para_treesort.tre", "nexus")

# # this will make the tree have node names labeled
# for node in tree.find_clades():
#     if node.confidence is not None and node.name is None:
#         node.name = node.confidence
        
fig = plt.figure(figsize=(15, 10)) # create figure & set the size 
matplotlib.rc('font', size=12)             # fontsize of the leaf and node labels 
matplotlib.rc('xtick', labelsize=10)       # fontsize of the tick labels

axes = fig.add_subplot(1, 1, 1)

axes.set_yticks([])  
axes.set_yticklabels([])  
axes.spines['left'].set_visible(False) 
axes.spines['right'].set_visible(False)
axes.spines['top'].set_visible(False)
axes.yaxis.label.set_visible(False) 
axes.title.set_visible(False)

tree.ladderize()
Phylo.draw(tree, show_confidence=False, axes=axes, label_func=lambda x: None)
```
