```python
import pandas as pd
import os
import re
import sys
import subprocess

module_dir = "/Users/monclalab1/Documents/scripts/"
sys.path.append(module_dir)

from fasta_editing import fasta_to_df, fasta_writer

list_of_genes = ["ha", "pb2","pb1","na","np","pa","ns","mp"]
```

```python
# write a loop that runs treesort using the output from the preceding run as the backbone
# if its on the first iteration of the for loop, just use h3nx_ha.fasta
# output should be iteratively named based on the for loop (output1.tre, output2.tre, etc)
# outputs should be put in a new folder called "outputs" (make this folder)
# descriptor.csv should be updated at the beginning of each run
# need to convert nexus treesort output to nwk
# add in augur traits snakefile
```

```python
if not os.path.exists("prepped/div/outputs"):
    os.makedirs("prepped/div/outputs")
else:
    pass

for i in range(1, 4):
    
    descriptor_entries = [] 
    
    for gene in list_of_genes:
        
        gene_label = f"*{gene.upper()}" if gene == "ha" else gene.upper()
        align_path = f"prepped/alignments/h3nx_{gene}.fasta"
    
        if gene == "ha":
            
            backbone_path = (
                "prepped/div/h3nx_ha.nwk" if i == 1 
                else f"prepped/div/outputs/output{i-1}.tree"
            )
            
            descriptor_entries.append([gene_label, align_path, backbone_path])
            
        else:
            div_path = f"prepped/div/h3nx_{gene}.nwk"
            descriptor_entries.append([gene_label, align_path, div_path])

    # descriptor csv
    with open("descriptor.csv", 'w') as descriptor_file:
        for row in descriptor_entries:
            descriptor_file.write(','.join(row) + '\n')
            
            
    if i == 1:

        treesort_cmd = [

            "treesort",
            "-i", "descriptor.csv",
            "-o", f"prepped/div/outputs/output{i}.tree"

            ]
    
    else:
        
        # need to convert treesort output to nwk form or treesort cant read it in lol
        
        with open(f"prepped/div/outputs/output{i-1}.tree", 'r') as file:
            nexus = file.read()
        
        start_idx = nexus.find('(')
        modified = nexus[start_idx:]

        end_idx = modified.find('END;')
        modified = modified[:end_idx]

        with open(f"prepped/div/outputs/output{i-1}.tree", "w") as output_file:
            output_file.write(modified.strip())
        

        treesort_cmd = [

            "treesort",
            "-i", "descriptor.csv",
            "-o", f"prepped/div/outputs/output{i}.tree",
            "--no-collapse"

            ]
        
    subprocess.run(treesort_cmd, check=True, capture_output=True, text=True)
```
