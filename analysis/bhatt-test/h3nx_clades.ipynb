# Subsetting code

```python
#this converts fasta files into dataframes

def fasta_to_df(fasta_file):
    
    fasta_data = []
    
    with open(fasta_file) as f:
        header = ""
        sequence = ""
        for line in f:
            if line.startswith(">"):
                if header != "":
                    fasta_data.append({"header": header, "sequence": sequence})
                header = line.strip() 
                sequence = ""
            else:
                sequence += line.strip()
        fasta_data.append({"header": header, "sequence": sequence}) #last line
            
    return pd.DataFrame(fasta_data)
```

```python
def fasta_writer(path, filename, df):
            
    try:  
        os.mkdir(path)

    except OSError as error:
        pass

    with open(f"{path}{filename}", "w") as f:
        for index, row in df.iterrows():
            f.write(f"{row['header']}\n")
            f.write(f"{row['sequence']}\n")
```

```python
def host_subsetting(gene, h3n2_list, h3n8_list):

    #subsetting the existing fasta file to only have the strains from the specific subtype
    df = fasta_to_df(f"/Users/monclalab1/Documents/nonhuman_H3_project/non-human-h3/sequences/h3nx_{gene}.fasta")
    
    df['Strain'] = df['header'].str.split("|").str[0]
    df['Strain'] = df['Strain'].str[1:]

    for key, value in h3n2_list.items():
        subset_df = df[df["Strain"].isin(value)]
        
        fasta_writer(f"/Users/monclalab1/Documents/nonhuman_H3_project/non-human-h3/analysis/host_specific/04-2023_11-2023/sep_by_clade/h3n2/{key}/", f"h3nx_{gene}.fasta", subset_df)

    for key, value in h3n8_list.items():
        subset_df = df[df["Strain"].isin(value)]

        fasta_writer(f"/Users/monclalab1/Documents/nonhuman_H3_project/non-human-h3/analysis/host_specific/04-2023_11-2023/sep_by_clade/h3n8/{key}/", f"h3nx_{gene}.fasta", subset_df)
```
